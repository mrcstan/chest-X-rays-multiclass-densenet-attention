# Append non-image features to training/validation/test lists that do not already contain the features

```python
import pandas as pd
import numpy as np
import os
from os import path
path_join = path.join
```

```python
BASE_PATH = "./labels"
input_files = ['train_A_0', 'train_A_1', 'train_A_2', 'train_A_3', 'train_A_4',
               'val_A_0', 'val_A_1', 'val_A_2', 'val_A_3', 'val_A_4',
               'train_val_A', 'test_A']
ext = '.csv'
out_postfix = 'nonImIn'

master_file = 'Data_Entry_2017_v2020.csv'


full_path = path_join(BASE_PATH, master_file)
dat_all = pd.read_csv(full_path)
dat_all = dat_all[['Image Index', 'Finding Labels','Patient Age', 'Patient Gender', 'View Position']]
dat_all.rename(columns={'Image Index': 'Image', 'Finding Labels': 'Labels'}, inplace=True)

for file in input_files:
    full_path = path_join(BASE_PATH, file+ext)
    print(full_path)
    dat = pd.read_csv(full_path)
   
    print('Before merging, ...')
    print(dat.head())
    dat = dat.merge(dat_all, on=['Image', 'Labels'])
    print('After merging, ...')
    print(dat.head())
    full_path = path_join(BASE_PATH, file+'_'+out_postfix+'.csv')
    dat.to_csv(full_path, index=False)

```
